```python
X = df.drop(columns = ['cardio'], axis = 1)
y = df['cardio']
```

```python
y = pd.DataFrame(y)
y
```

```python
#importing necessary librearies for machine learning
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score, f1_score
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state = 350)
```

```python
#Implementing LogisticRegression

log_reg = LogisticRegression()

log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
#accuracy score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
conf_mat
```

```python
# F1 score
score = f1_score(y_test, y_pred)
score
```

```python
#Area under curve
auc = roc_auc_score(y_test, y_pred)
auc
```

```python
#Implementing Support Vector Machine

from sklearn.svm import SVC
```

```python
clf = SVC()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
#Implementing Decision Tree

from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train, y_train)
```

```python
dt_clf.score(x_test, y_test)
```

```python
y_pred = dt_clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred) 
accuracy
```

```python
#Implementing KNN

from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```

```python
#Implementing Random Forest

from sklearn.ensemble import RandomForestClassifier
```

```python
rand_clf = RandomForestClassifier()

rand_clf.fit(x_train, y_train)
rand_clf.score(x_test, y_test)
```

```python
grid_param = {
    "n_estimators" : [90, 100, 115],
    'criterion' : ['gini', 'entropy'],
    'min_samples_leaf' : [1,2,3,4,5],
    'min_samples_split' : [4,5,6,7,8],
    'max_features' : ['auto', 'log2']
}
```

```python
grid_search = GridSearchCV(estimator = rand_clf, param_grid = grid_param, cv = 5, n_jobs = -1)
```

```python
grid_search.fit(x_test, y_test)
```

```python
grid_search.best_params_
```

```python
rand_clf = RandomForestClassifier(criterion = 'entropy', max_features = 'auto', min_samples_leaf = 5, min_samples_split = 4, n_estimators = 115)
```

```python
rand_clf.fit(x_test, y_test)
```

```python
rand_clf.score(x_test, y_test)
```
